```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
datos_exportaciones = pd.read_csv("C:/Users/marco/Desktop/Desarrollos/Analisis y visualizacion de datos/ARCHIVOS/exportaciones.csv", encoding="latin-1")
```

```python
datos_total_pais = datos_exportaciones[(datos_exportaciones['provincia'] == 'Total Pas') & (datos_exportaciones['rubro'] == 'Total')]

plt.figure(figsize=(12, 6))
plt.bar(datos_total_pais['anio'], datos_total_pais['value'], label='Total Pas')
plt.xlabel('Ao')
plt.ylabel('Total')
plt.title('Exportaciones')
plt.legend()
plt.grid(True)
plt.show()
```
