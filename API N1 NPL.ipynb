   A) El analisis de sentimientos es una aplicacion que permite identificar y extraer opiniones subjetivas de textos. Se utiliza para comprender mejor las percepciones de los clientes sobre servicios, productos, marcas, etc. Esto se realiza a traves del analisis de los comentarios, reseñas o publicaciones. Esta informacion nos va a servir para mejor la experiencia para el usuario y/o estrategia de marketing.
   
   B) Los pasos para convertir los comentarios en datos esctructurados son:
   1. Se toman los comentarios originales
   2. Limpieza y Normalizacion: Se normaliza el texto eliminando los caracteres especiales, innecesarios, mala ortografia, etc.
   3. Tokenizacion: Dividimos el texto en palabras o tokens
   4. Reduccion: Se reducen las palabras eliminando sufijos y prefijos innecesarios
   5. Transformacion: Se logra una representacion matematica del texto a partir del vocabulario definido
